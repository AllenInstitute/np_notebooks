````python
df['sweep_frames']
````

- each stimulus has a `sweep_frames` list
- denotes "display times" for each frame of the stimulus, expressed as frame idx relative to
  the start of the stimulus
  - frame index updates at 60 fps 
  - each frame of a stimuls is displayed for two screen flips: effectively 30 fps 
- each `sweep_frames` list has sufficient indices to loop the stimulus 10 times, according to `runs` (will make more sense when we
  look at `sweep_order` below)
- `len_stim = len(sweep_frames) / len(runs)` 

````python
fig, ax = plt.subplots(3, 1, sharex=True, sharey=True, figsize=(8,6))
for _ in range(3):

    stim = df.iloc[stim_epocs[_]]
    
    x = []
    for frames in stim['sweep_frames']:
        x.extend(frames)

    ax[_].plot(np.array(x)  , np.zeros_like(x))
    ax[_].set(title=stim_epoc_names[_], xticks=(0, len(x)))
    if _ < 2:
        ax[_].annotate(f'{len(x)}', (len(x), 0), xytext=(0, 10), textcoords='offset points', ha='center', arrowprops=dict(arrowstyle='->'))
    
ax[_].set(xlabel='frames');
````

# Recreate missing fields

These fields aren't saved in every .pkl file, since they're the same for each experiment
and are quite large: `sweep_frames`, `sweep_order`, `frame_list`

To recreate them for any stim in sequence [0:135]:

````python
stim_idx = 34

stim = df.loc[stim_idx]

# >>> stim.display_sequence
# array([[1360, 1400]])

stim_frames = int((stim.display_sequence[0][-1] - stim.display_sequence[0][0]) * stim.fps)
flips_per_frame = int(stim.fps * stim.frame_length)
````

````python
stim
````

````python
sweep_frames = []
for n in range(int(stim.runs)):
    sweep_frames.extend([(_, _ + 1) for _ in range(stim_frames * n, stim_frames * (n + 1), flips_per_frame)])

assert sweep_frames == stim.sweep_frames
````

````python
sweep_order = [int(_) for _ in np.mod(np.arange(len(sweep_frames)), stim_frames / flips_per_frame)]

assert sweep_order == stim.sweep_order
````

````python
frame_list = list(-1 * np.ones(int(stim.display_sequence[0][0] * stim.fps), dtype=int))

assert (a := len(frame_list)) == (b := len(stim.frame_list[stim.frame_list == -1])), f'{a} != {b}'

for idx, _ in enumerate(sweep_frames[:int(stim_frames / flips_per_frame)]):
    frame_list.extend([sweep_order[idx]] * flips_per_frame)
    
frame_list = np.array(frame_list)
        
assert all(np.equal(frame_list, stim.frame_list))
````
